```python

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns
```

```python
data_path = pd.read_excel("D:\jena_climate_2009_2016(2)1 .xlsx")
data_path.head()
```

```python
plt.figure(figsize=(16,8))
rc = {'font.sans-serif': 'SimHei',
      'axes.unicode_minus': False}
sns.set(context='notebook', style='ticks', rc=rc)
sns.lineplot(x="气压(毫巴，百帕)",y="温度 (摄氏度)",data=data_path[:10000])
plt.show()

```

```python
plt.figure(figsize=(26,16))
sns.pointplot(x='日',y='温度 (摄氏度)',data=data_path[0:50000],hue='月')
rc = {'font.sans-serif': 'SimHei',
      'axes.unicode_minus': False}
plt.show()
```

```python
plt.figure(figsize=(16,8))
rc = {'font.sans-serif': 'SimHei',
      'axes.unicode_minus': False}
sns.set(context='notebook', style='ticks', rc=rc)
sns.lineplot(x="温度 (摄氏度)",y="相对于湿度的摄氏温度。露点是空气中水的绝对量的量度，是空气不能保持其中所有水分并且水凝结的温度。"
             ,data=data_path[:10000])
plt.show()
```

```python
plt.figure(figsize=(16,8))
rc = {'font.sans-serif': 'SimHei',
      'axes.unicode_minus': False}
sns.set(context='notebook', style='ticks', rc=rc)
sns.lineplot(x="温度 (摄氏度)",y="风速(m/s)",data=data_path[:10000])
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.boxplot(x='月',y="温度 (摄氏度)", data=data_path)
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.relplot(x="温度 (摄氏度)",y="年",data=data_path)
plt.show()
```
